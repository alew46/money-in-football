# Analyzing The Effect Of Money In Football

The 1995 [Bosman ruling](https://en.wikipedia.org/wiki/Bosman_ruling), in which the European Court of Justice effectively removed contracuatual limits on football clubs spending huge sums to secure the services of superstar players, has had a massive effect on the sport of football and indeed in all of global sport.

Jean-Marc Bosman was an otherwise unremarkable football player, even by a generous interpretation of the word. But his legal challenge to his club's refusal to release him from an expired contract granted individual footballers (and their agents) the power to move to any club, in any country, for whatever fee the purchasing club deemed fitting. Thus the "open market" system was born.

But the open market system is only one part of the story when it comes to modern football business. In the mid-2000s, investment in clubs by both hyper wealthy individuals and conglomorated ownership businesses ramped up significantly. High profile owners from Russia, the United States, and the United Arab Emirates purchased historic football clubs and funneled huge sums via both direct investment and indirect means such as sponsorships, eventually giving rise to the present-day trend of juggernaut clubs like Bayern Munich and Real Madrid consistently engaging in bidding wars for the world's most prestigious players.

The backlash against this money-powered world of football has been, paradoxically, both severe and toothless. While everyday football fans often decry the ability of clubs to "buy" their way to success, attempts by authorities to crack down on opulent spending through mechanisms like the [Financial Fair Play rules](https://en.wikipedia.org/wiki/UEFA_Financial_Fair_Play_Regulations) have been widely lauded as ineffective and even corrupt.

Questions abound in the world of money in football. Is it really the case that only big-money clubs with superstar players can win? Where does all this money come from - both geographically and in terms of business? Are the "corporate" clubs the only successful clubs? 

And perhaps most importantly: in a sport where everyone loves a plucky underdog story, can clubs with less means ever compete again? Or are we doomed to forever watch the same handful of clubs vie for their domestic leagues and international cups, while the have-nots languish in the lower table?

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
pd.options.display.float_format = '{:.0f}'.format

transfers = pd.read_csv('data/top250-00-19.csv')
clubs = pd.read_csv('data/club_value_ownership.csv')
```

#### Clubs are spending more on star players than ever before.

A look at average transfer spending across the wealthiest clubs in the world shows a definite upward trend since the late 2000's, with a more marked increase starting around 2013.

```python
avg_annual_spend = transfers.groupby('Season').agg({'Transfer_fee':np.mean}).reset_index()
top_ten_transfers = transfers.sort_values('Transfer_fee',ascending = False).head(10)

np.random.seed(42)
cmap = matplotlib.cm.get_cmap('Spectral')
color_dict = pd.Series({k:cmap(np.random.rand()) for k in top_ten_transfers['League_to'].unique()})
color_dict.name = 'color_dict'
top_ten_transfers = pd.merge(top_ten_transfers, color_dict, how='left', left_on='League_to', right_index=True)
plt.rcParams["figure.figsize"] = [12, 6]

_ = avg_annual_spend.plot.bar(x='Season', y='Transfer_fee')
```

This must mean something besides the Bosman ruling is at work here. If it was simply the case that the open market had freed up big spending clubs to pay huge sums for players, then we'd expect to see a steady increase in money spent, with prices for players rising but ultimately constrained the same way any other commodity is.

While it's true that there's more money than ever in football, it could also be the case that available funds have simply moved hands - from many clubs, to comparatively fewer clubs - and concentrated in the hands of the wealthiest. Let's dig in to some mor details about club spending to see.

#### Spending is on the rise across the globe - but not every club in every country is taking part.

```python
epl_clubs = clubs[clubs['competition'] == 'Premier League']
laliga_clubs = clubs[clubs['competition'] == 'LaLiga']
bundesliga_clubs = clubs[clubs['competition'] == 'Bundesliga']

_ = epl_clubs.hist(column='clubMarketValue', figsize=(5,5), bins=5)
_ = plt.title('Wealth Distribution in the Premier League')

_ = laliga_clubs.hist(column='clubMarketValue', figsize=(5,5), bins=5)
_ =plt.title('Wealth Distribution in La Liga')

_ = bundesliga_clubs.hist(column='clubMarketValue', figsize=(5,5), bins=5)
_ = plt.title('Wealth Distribution in the Bundesliga')
```

The data make it clear - though global spend is on the rise, it's only the wealthiest handful of clubs who could possibly be driving the trend. In the English Premier League and Spanish La Liga, there are clearly two groups of clubs; one bottom-heavy group of have nots where there are far more clubs towards the bottom of the barrel than the middle, and one group of elites with much higher valuations than their peers.

The German Bundesliga, perhaps befitting its name, initially looks more equitable. However, it may have an even more entrenched group of ultra-wealthy outliers, with just one club each occupying each of the wealth bands above the pack.

#### Does more money translate to more wins?

Conventional wisdom would say that, in a post Bosman world, the clubs that can afford to pay the most for players will win more than others. Each club can only field 11 players at a time, but the rich clubs can make sure the 11 they field are among the best in the world at their respective positions, as determined by the free market value of those players.

Let's look at the correlation between club value and number of top flight wins for what is typically regarded as one of the most competitive leagues in the world, the English Premier League.

```python
_ = epl_clubs.plot.scatter(x='clubMarketValue', y='topFlightWins', xlabel='Club Value', ylabel="Top Flight Wins", s=epl_clubs['clubMarketValue'], alpha=.8)
```

Sure enough, we can see the richest clubs, represented by the larger bubbles and the more rightward positions on the plot, tend to take home the trophies. 

Even some of the departures from this trend The richest club, Manchester City, has a relatively modest 8 league wins - on par with some much poorer teams. However, their wins have been dramatically concentrated into recent years, after their large infusion of wealth via an ownership swap in the last decade. 6 of their 8 championships occurred after 2011.

#### Where does the money come from?

We've established that the mass influx of money into the global football market is a relatively new phenomenon, even by post Bosman standards. And we've seen that it's having a real, stratifying impact on who takes home the trophies in some of the biggest leagues in the world.

But where does all this money come from? Who are the ones doing the buying and funding these teams, and from where do they hail?

Both of these owner types have an outsize amount of team value relative to their number.

Notably, there is one club - Newcastle United - that was recently purchased by a soverign wealth entity powered by the Saudi royal family. It is generally accepted that this is one of the richest families in the world, with virtually unlimited resources at their disposal. However, Newcastle's market value hasn't risen to the levels of the EPL elite. Why?

Is this the effect of Financial Fair Play rules at work, which are meant to limit the amount of money that owners can directly funnel to their club's transfer warchest? Is it only a matter of time before Newcastle's value increases, as the owners leverage their resources to find indirect ways of funding their teams i.e. via sponsorships and marketing? Time will surely tell.

```python
epl_clubs.replace('Welsh', 'United Kingdom', inplace=True)
```

```python
#function to convert to alpah2 country codes and continents
from pycountry_convert import country_alpha2_to_continent_code, country_name_to_country_alpha2
def get_continent(col):
    try:
        cn_a2_code =  country_name_to_country_alpha2(col)
    except:
        cn_a2_code = 'Unknown' 
    try:
        cn_continent = country_alpha2_to_continent_code(cn_a2_code)
    except:
        cn_continent = 'Unknown' 
    return (cn_a2_code, cn_continent)
```

```python
countries = {}
continents = {}

for index, row in epl_clubs.iterrows():
    country, continent = get_continent(row.ownershipCountry)
    countries[row.club] = country
    continents[row.club] = continent
```

```python
countries_df = pd.DataFrame(list(countries.items()), columns = ['club','Country'])
continents_df = pd.DataFrame(list(continents.items()), columns = ['club','Continent'])

epl_clubs = pd.merge(epl_clubs, countries_df, on='club')
epl_clubs = pd.merge(epl_clubs, continents_df, on='club')
```
